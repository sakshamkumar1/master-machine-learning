### Heart disease frequency according to sex

```python
df["sex"].value_counts()
```

```python
df[df["sex"] == 1]["target"].value_counts()
```

```python
df[df["sex"] == 0]["target"].value_counts()
```

```python
pd.crosstab(df["target"], df["sex"])
```

```python
ax = pd.crosstab(df["target"], df["sex"]).plot(kind="bar",
                                               figsize=(10, 6));

ax.set(title="Heart Disease with Sex", xlabel="Target",
       ylabel="Frequency of Heart Disease");
ax.legend(["Female", "Male"]);
```
